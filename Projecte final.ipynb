```python
def hora_punta(hora):
    if (7 <= hora < 9) or (13 <= hora < 15) or (17 <= hora < 19):
        return 'Hora punta'
    else:
        return 'Hora no punta'

accidents['Hora_Punta'] = accidents['Hora_dia'].apply(hora_punta)

plt.figure(figsize=(10, 6))
sns.countplot(x='Hora_Punta', data=accidents, palette='coolwarm')
plt.title("Comparació d'accidents entre hora punta i no punta")
plt.xlabel('Hora')
plt.ylabel("Quanitat d'accidents")
plt.show()
```

* Ara faré algun test estadistic per a estudiar la correlació amb més profunditat. Per això, es necessari verificar que la meva mostra segueix una distribució normal. Per tant, el primer pas sera verificar que la disgribuciño de les meves dades es normal. Per a fer-ho, procedeixo a realitzra probes de normalitat:

```python
from scipy import stats
stat, critical_values, significance_level = stats.anderson(accidents['Hora_dia'], dist='norm')
print(f"Estadístico de Anderson-Darling: {stat}")
print(f"Valores críticos: {critical_values}")
print(f"Niveles de significancia: {significance_level}")
```

```python
# Test estadisrtic per a mostres que no segueixen una distribució normal
from scipy.stats import mannwhitneyu

# Agafo hores punta i n punta
hora_punta = accidents[accidents['Hora_Punta'] == 'Hora punta']
hora_no_punta = accidents[accidents['Hora_Punta'] == 'Hora no punta']

accidentes_punta = hora_punta.groupby('Hora_Punta').size()
accidentes_no_punta = hora_no_punta.groupby('Hora_Punta').size()

# Prova de Mann-Whitney U
stat, p_value = mannwhitneyu(accidentes_punta, accidentes_no_punta)

print(f"Estadístico U: {stat}")
print(f"P-valor: {p_value}")

# P-valor
if p_value < 0.05:
    print("Es rebutja la hipótesis nula. Hi ha una diferencia significativa entre horas punta i no punta.")
else:
    print("No es rebutja la hipótesis nula. No hi ha una diferencia significativa entre horas punta i no punta")


# H₀: No hi ha una diferencia significativa entre horas punta i no punta.
# H₁: hi ha una diferencia significativa entre horas punta i no punta.
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Asumiendo que el dataframe 'accidents' ya contiene una columna llamada 'Hora_dia'
# Función para clasificar si es hora punta o no
def hora_punta(hora):
    if (7 <= hora < 9) or (13 <= hora < 15) or (17 <= hora < 19):
        return 'Hora punta'
    else:
        return 'Hora no punta'

# Aplicar la función al dataframe
accidents['Hora_Punta'] = accidents['Hora_dia'].apply(hora_punta)

# Separar los datos de horas punta y no punta
hora_punta = accidents[accidents['Hora_Punta'] == 'Hora punta']['Hora_dia']
hora_no_punta = accidents[accidents['Hora_Punta'] == 'Hora no punta']['Hora_dia']

# Probar si las distribuciones de horas de accidentes son significativamente diferentes
stat, p_value = mannwhitneyu(hora_punta, hora_no_punta)

# Resultados
print(f"Estadístico U: {stat}")
print(f"P-valor: {p_value}")

# Interpretación del p-valor
if p_value < 0.05:
    print("Es rebutja la hipótesis nula. Hi ha una diferencia significativa entre hores punta i no punta.")
else:
    print("No es rebutja la hipótesis nula. No hi ha una diferencia significativa entre hores punta i no punta")

# Hipótesis:
# H₀: No hi ha una diferència significativa entre hores punta i no punta.
# H₁: Hi ha una diferència significativa entre hores punta i no punta.
```
